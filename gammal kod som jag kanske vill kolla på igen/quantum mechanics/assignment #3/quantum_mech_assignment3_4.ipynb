```python
import numpy as np
from matplotlib import pyplot as plt

### define some constants ###
hbar = 1
c = 1
q = 1
rho = 1
m = 1

### define energy as function of B0 ###
def E_l(B0, l=0):
    return (l - q * B0 * rho**2 / (2 * hbar * c))**2 * hbar**2 / (2 * m * rho**2)

### plot energy for different l ###
l_vec = range(-8, 8)
B0_vec = np.linspace(-32., 32., 100)
fig, ax = plt.subplots(figsize=(12,8))
plasma = plt.cm.plasma(np.linspace(0, 1, len(l_vec))) # color for plot

for l in l_vec:
    ax.plot(B0_vec, E_l(B0_vec, l=l), color=plasma[l], label=f'l = {l}')

ax.set_xlabel(r'B_0', fontsize=24)
ax.set_ylabel(r'E_l', fontsize=24)
ax.set_xticks([])
ax.set_yticks([])
ax.legend(fontsize=12)
plt.show()
```
